Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

Data Description: Murder -- Muder rates in different places of United States Assualt- Assualt rate in different places of United States UrbanPop - urban population in different places of United States Rape - Rape rate in different places of United States

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
crime=pd.read_csv("/content/crime_data.csv")
```

```python
crime
```

```python
#Normalise the data
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
crime1=norm_func(crime.iloc[:,1:])
```

```python
crime1
```

```python
#create Dendogram
dendogrm=sch.dendrogram(sch.linkage(crime1,method="average"))
```

```python
# Create Clusters define Agglomerative
hc=AgglomerativeClustering(n_clusters=4,affinity="euclidean",linkage="average")
```

```python
# Fit the data in the model use fit_predict
y_hc=hc.fit_predict(crime1)
Clusters=pd.DataFrame(y_hc,columns=["clusters"])
```

```python
Clusters
```

```python
crime["clusterid"]=Clusters
```

```python
crime.groupby("clusterid").agg(["mean"]).reset_index()
```

```python
# Here we can see 0 cluster has lowest crime rate compared to others

crime[crime["clusterid"]==0]
```

```python
plt.figure(figsize=(10,7))
plt.scatter(crime["clusterid"],crime["Murder"],c=hc.labels_)
```
